```python
# 가독성을 높히기 위해 만원 단위로 나누기
df['월관리비'] = df['월관리비'] / 10000
df['보증금/전세금(원)'] = df['보증금/전세금(원)'] / 10000
df['월세(원)'] = df['월세(원)'] / 10000
df['매매가(원)'] = df['매매가(원)'] / 10000

# 칼럼명에 (만원) 단위 표시를 위해 칼럼명 수정
df.rename(columns={'월관리비':'월관리비(만원)'}, inplace=True)
df.rename(columns={'보증금/전세금(원)':'보증금/전세금(만원)'}, inplace=True)
df.rename(columns={'월세(원)':'월세(만원)'}, inplace=True)
df.rename(columns={'매매가(원)':'매매가(만원)'}, inplace=True)
df
```

### 개포동 내에서 가장 매물이 많은 아파트 단지 개포래미안포레스트와 개포자이프레지던스의 전세가율/갭
- 개포래미안포레스트
    - 평균 전세가율 (평균 전세가 / 평균 매매가 * 100%)
        - 12억 5544만원 / 26억 3012만원 * 100% = 47.73%
    - 갭
        - 26억 3012만원 - 12억 5544만원 = 13억7468만원

- 개포자이프레지던스
    - 평균 전세가율
        - 15억9023만원 / 55억1741만원 * 100% = 28.82%
    - 갭
        - 55억1741만원 - 15억9023만원 = 39억2718만원
